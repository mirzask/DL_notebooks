```python
bs = 64
data = ImageDataBunch.from_folder(path, train='train', valid = 'val', ds_tfms=get_transforms(max_rotate=5.0, max_warp=0),
                                  size=224, bs=bs).normalize(imagenet_stats)
```

```python
print(data.classes)
len(data.classes),data.c
```

```python
data
```

# Fit the model

Using `ResNet34`

```python
kappa = KappaScore()
kappa.weights = "quadratic"
```

```python
learn = cnn_learner(data, models.resnet34, metrics=[error_rate, accuracy, kappa], 
                    wd=0.1,
                    callback_fns=ShowGraph)
```

```python
learn.model
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 4e-3
learn.fit_one_cycle(10,max_lr=lr)
```

# How's it looking?

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
x,y = data.valid_ds[3]
x.show()
data.valid_ds.y[3]
```

```python
print(type(learn))
print(type(x))
```

```python
gcam = GradCam.from_one_img(learn, x)
gcam.plot()
```

Here, we see that our model predicts that this will be KL grade 3 with probability 0.993, which is the correct category for this particular X-ray. The Grad-CAM also appears to highlight the joint space narrowing, osteophytes, and other changes that warrant this categorization.

```python
interp.plot_top_losses(9, figsize=(15,15))
```

> It may just be me, but I feel like these poor performers appear to have low contrast/brightness in common.

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

# Using RAdam optimizer

```python
from radam import *
```

```python
optar = partial(RAdam)
```

```python
learn = cnn_learner(data, models.resnet34, metrics=[error_rate, accuracy, kappa], 
                    wd=1e-3,
                    opt_func=optar,
                    callback_fns=ShowGraph)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 4e-3
learn.fit_one_cycle(10,max_lr=lr)
```

# Inspection

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
gcam = GradCam.from_one_img(learn, x)
gcam.plot()
```

```python
interp.plot_top_losses(9, figsize=(15,15))
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```
